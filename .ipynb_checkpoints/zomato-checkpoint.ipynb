```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv(r"zomato.csv",encoding='latin-1')
```

```python
df
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
# working with missing values
df.isnull().sum()
```

```python
plt.figure(figsize=(12,9))
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

Only cuisines have 9 missing values

```python
df_country = pd.read_excel("Country-Code.xlsx")
```

```python
df_country
```

```python
final_df = pd.merge(df, df_country, on="Country Code", how="left")
```

```python
country_names = final_df["Country"].value_counts().index
```

```python
country_values = final_df["Country"].value_counts().values
```

```python
plt.figure(figsize=(12,6))

plt.pie(country_values[:3], labels=country_names[:3], autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
# understanding the rating prespective
rating_df = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index()
rating_df
```

```python
rating_df = rating_df.rename(columns={0:'Rating count'})
```

```python
rating_df
```

Obsering the Ratings: <br>
When rating is 1.8 to 2.4 it is Poor <br>
When rating is 2.5 to 3.4 it is Average<br>
When rating is 3.5 to 3.9 it is Good<br>
When rating is 4.0 to 4.4 it is Very Good<br>
When rating is 4.5 to 4.9 it is Excellent<br>

```python
plt.figure(figsize=(12,9))
sns.barplot(x='Aggregate rating', y='Rating count', data=rating_df)
```

```python
plt.figure(figsize=(12,9))
sns.barplot(x='Aggregate rating', y='Rating count', hue='Rating color',data=rating_df, palette=['blue','red','orange','yellow','green','green'])
```

# Observation
Not Rated count is Very High as we can see in blue.

Maximum number of ratings are between 2.5 to 4.1

```python
sns.countplot(x="Rating color", data=rating_df, palette=['blue','red','orange','yellow','green','green'])
```
